```python
def extract_game_logs(name, player_id, season_year):
    url = f"https://www.espn.com/mens-college-basketball/player/gamelog/_/id/{player_id}/season/{season_year}"
    response = requests.get(url, headers=headers)
    match = re.search(r"window\['__espnfitt__'\]\s*=\s*({.*?});\s*</script>", response.text, re.DOTALL)
    if not match:
        return None

    try:
        parsed_json = json.loads(match.group(1))
        gmlog = parsed_json["page"]["content"]["player"]["gmlog"]
        stat_abbrs = [gl["abbr"] for gl in gmlog.get("gls", [])]
        logs = []
        for group in gmlog.get("groups", []):
            for tbl in group.get("tbls", []):
                for event in tbl.get("events", []):
                    stats = dict(zip(stat_abbrs, event["stats"]))
                    logs.append({
                        "date": event["dt"][:10],
                        "opponent": event["opp"]["name"],
                        "result": event["res"]["abbr"],
                        "score": event["res"]["score"],
                        "context": event.get("nt", ""),
                        "stats": stats
                    })
        return logs if logs else None
    except Exception as e:
        return None

# Iterate over seasons until no player returns data
season_year = 2025
all_data = {}

while True:
    print(f"Scraping season: {season_year}")
    season_data = {}
    found_data = False

    for name, pid in players.items():
        logs = extract_game_logs(name, pid, season_year)
        if logs:
            found_data = True
            season_data[name] = logs

    if not found_data:
        print(f"No data found for season {season_year}. Stopping.")
        break

    all_data[season_year] = season_data
    season_year -= 1

# Optional: Save to JSON file
with open("all_game_logs.json", "w") as f:
    json.dump(all_data, f, indent=2)
```
